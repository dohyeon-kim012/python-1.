# 1. Scatter Plot

## 1-1. x, y, colors, area 설정하기
* colors : 임의의 값을 색상으로 표현
* area : 점의 크기. 값이 커지면 점도 커짐

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.arange(50)
area = x * y * 1000
```

```python
plt.scatter(x, y, s=area, c=colors)  # s : 점의 크기, c : 점의 색상
plt.show()
```

## 1-2. cmap, alpha
* cmap에 컬러를 지정하면, 컬러 값을 모두 동일하게 설정
* alpha는 투명도를 나타내며, 0~1사이의 값을 갖고, 0에 가까울 수록 투명해진다.

```python
# 현재 사용할 figure에 대해서만 크기 조절하기
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='Blues', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(132)
plt.scatter(x, y, s=area, cmap='Plasma', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, c=y, alpha=1.0)
plt.title('alpha=1.0, colorbar')
plt.colorbar()  # 값별 농도 확인하기

plt.show()
```

# 2. Bar Plot, Barh Plot


## 2-1. 기본 Bar Plot 그리기

```python
# x에 들어가는 카테고리만큼 y 에도 값이 들어가야 함
x = ['Math', 'Programming', 'Science', 'English', 'PE']
y = [70, 80, 60, 75, 40]

plt.bar(x, y)
###
plt.xticks(x)  # xticks 옵션의 중요성: 카테고리의 설정을 의미하기 때문에
###
plt.xlabel('Subjects')
plt.ylabel('Score')

plt.show()
```

## [번외] Barplot에서 비교 그래프 그리기
* 하나의 캔버스에 2개 이상의 Bar 그리기

```python
x_label = ["Math", "Programming", "Science", "English", "PE"]
x = np.arange(len(x_label)) # 실제 label의 개수만큼 정수 배열 생성

y_1 = [70, 80, 60, 40, 50] # 첫 번째 사람의 점수
y_2 = [100, 60, 90, 45, 90] # 두 번째 사람의 점수

fig, axes = plt.subplots() # 1, 1 형태의 ax 하나 생성.

# bar의 넓이 지정 (왼쪽을 지정할 때는 - width/2 , 오른쪽은 + width/2)
width = 0.35
axes.bar(x - width/2, y_1, width, align='center', alpha=0.5)
axes.bar(x + width/2, y_2, width, align='center', alpha=0.8)

# xticks 설정
plt.xticks(x)

axes.set_xticklabels(x_label) # 실제 표시할 레이블

plt.show()
```

# 3. Line Plot

## 3-1. 기본 lineplot 그리기

```python
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.grid()  # grid 표시
plt.show()
```

## 3-2. 2개 이상의 그래프 그리기
* color, alpha 추가

```python
# ★★★line 그래프에서: x의 값이 기준이기 때문에 변하면 안 됨!!!★★★

x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, color='blue', alpha=0.3, label='1+sin')
plt.plot(x, y_2, color='red', alpha=0.7, label='1+cos')

plt.grid()
plt.legend()
plt.show()
```

* marker 추가

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, color='blue', alpha=0.3, label='1+sin', marker='o')
plt.plot(x, y_2, color='red', alpha=0.7, label='1+cos', marker='+')

plt.grid()
plt.legend()
plt.show()
```

* linestyle 변경

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, color='blue', alpha=0.3, label='1+sin', linestyle='--')
plt.plot(x, y_2, color='red', alpha=0.7, label='1+cos', linestyle='-.')

plt.grid()
plt.legend()
plt.show()
```

# 4. Area Plot

## 4-1. 기본 Areaplot

```python
x = np.arange(1, 21)  # 1 ~ 20 까지 정수 배열 생성
y = np.random.randint(low=5, high=10, size=20)

# fill_between 으로 색칠하기
# lineplot과 사용방법은 똑같으나, 밑을 채워주는 점이 차이점

plt.fill_between(x, y, color='skyblue', alpha=.4)
plt.show()
```

## 4-2. 경계선을 굵게, area는 옅게 표현해보기

```python
plt.fill_between(x, y, color='skyblue', alpha=0.3)
plt.plot(x, y, color='skyblue', alpha=0.8)
```

## 4-3. 여러 그래프를 겹쳐서 표현

```python
x = np.arange(1, 10, 0.05)  # np.arange = np.linspace
y_1 = np.cos(x) + 1
y_2 = np.sin(x) + 1
y_3 = y_1 * y_2 / np.pi

plt.fill_between(x, y_1, color='green', alpha=.1)
plt.fill_between(x, y_2, color='blue', alpha=.2)
plt.fill_between(x, y_3, color='red', alpha=.3)

plt.show()
```

# 5. Histogram

## 5-1. 기본 Histogram

```python
N = 100000
bins = 30  # bins -> 구간

x = np.random.randn(N)

plt.hist(x, bins=bins)  # y축이 필요 없음 : 히스토그램은 기본적으로 y축이 구간별 count를 의미
plt.show()
```

## 5-2. 다중 Histogram

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 3,
                         sharey=True,   # y축 그래프를 모든 다중 그래프가 공유
                         tight_layout=True)  # graph의 패딩을 자동으로 조절해서 fit한 graph를 생성

fig.set_size_inches(12, 5)

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins=bins * 2)
axes[2].hist(x, bins=bins * 4)

plt.show()
```

## 5-3. y축에 Density 표기

* 밀집도가 1이면 : "모든 데이터" 의미

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, 
                         tight_layout=True)

fig.set_size_inches(9, 3)

# density=True 값을 통하여 y축에 Density를 표기할 수 있다
axes[0].hist(x, bins=bins, density=True, cumulative=True)  # cumulative: 누적값
axes[1].hist(x, bins=bins, density=True)

plt.show()
```

# 6. Pie Chart
* `explode` : 파이에서 한 비율만 나와 보이게 하는 효과
* `autopct` : 퍼센트 자동으로 표기
* `shadow` : 파이차트에 그림자 효과
* `startangle` : 파이를 그리기 시작할 각도

pie 차트를 활용하면 2가지 리턴을 받음
* `texts` : label에 대한 텍스트 효과
* `autotexts` : 파이 위에 그려지는 텍스트 효과

```python
labels = ['A', 'B', 'C', 'D', 'E', 'F']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)

# texts, autotexts 인자를 활용하여 텍스트 스타일링 적용
_, texts, autotexts = plt.pie(sizes,
                              labels=labels,
                              explode=explode,
                              autopct='%1.1f%%',  # '%(패턴 시작)1(최소 한자리).1(최대 한자리)f% %(패턴 끝)'
                              shadow=True,
                              startangle=0)

plt.title('PIE CHART TEST')

# label 텍스트에 대한 스타일 적용
for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
    t.set_color("white")
    t.set_fontsize(18)

plt.show()
```

# 7. Box Plot

```python
# 기본 데이터 분포
spread = np.random.rand(50) * 100
center = np.ones(25) * 50  # 가운데에 값이 더 많아지도록 중간값을 추가함

# 윗쪽 아웃라이어 만들기
flier_high = np.random.rand(10) * 100 + 100  # 최소 100, 최대 200

# 아랫쪽 아웃라이어 만들기
flier_low = np.random.rand(10) * -100  # 최소 -100, 최대 0

# 모든 데이터 합치기
data = np.concatenate([spread, center, filer_high, flier_low])
len(data)
```

```python
plt.hist(data)
```

```python
plt.boxplot(data)
plt.show()
```

## 7-2 다중 박스플롯 생성

```python
spread = np.random.rand(50) * 100
center = np.ones(25) * 50  
flier_high = np.random.rand(10) * 100 + 100 
flier_low = np.random.rand(10) * -100  
data = np.concatenate([spread, center, filer_high, flier_low])

spread = np.random.randn(50) * 100
center = np.ones(25) * 40 # 중앙값을 약간 내려보기 (50 -> 40)
flier_high = np.random.rand(10) * 100 + 100 
flier_low = np.random.rand(10) * -100

d2 = np.concatenate([spread, center, flier_high, flier_low])
```

```python
data.shape, d2.shape
```

```python
data_re = data.reshape(-1, 1)
d2_re = d2.reshape(-1, 1)
```

```python
final_data = [data_re, d2_re, d2_re[::2, 0]]  # 이를 통해 3개의 boxplot을 만들 수 있음
```

```python
plt.boxplot(final_data)
plt.show()
```

`vert=False` 옵션을 통해 축 변경

```python
plt.boxplot(final_data, vert=False)
plt.show()
```

## 7-4. Outlier 마커 심볼, 컬러 변경

```python
outlier_marker = {"markerfacecolor":"r", "marker":"D"}
plt.boxplot(final_data, flierprops=outlier_marker)
plt.show()
```
