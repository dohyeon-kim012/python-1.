# 1. 밑 그림 그리기
* 그래프가 그려지는 공간 : canvas
* canvas를 다룰 수 있는 개체 : figure

## 1-1. 단일 그래프

````python
# 데이터 생성
data = np.arange(1, 100)

# 그리기(plot)
plt.plot(data)

# 그래프 보여주기
plt.show()
````

## 1-2. 다중 그래프 (multiple graphs)
1개의 canvas에 여러 개의 그래프 그리기

````python
data1 = np.arange(1, 51)
data2 = np.arange(51, 101)

# 캔버스에 그래프를 올려가는 과정
plt.plot(data1)
plt.plot(data2)

# show 전에 여러 개의 그래프를 동시에 올릴 수 있다
plt.show()
````

2개의 figure로 나누어서 다중 그래프 그리기
* `plt.figure()` : 캔버스 설정 함수, 호출 할 때마다 캔버스가 하나씩 만들어짐

````python
plt.plot(data1)  # data1 시각화

plt.figure()  # 새로운 캔버스 생성

plt.plot(data2)  # 새롭게 만들어진 캔버스에 data2 시각화

plt.show  # 모두 보여주기
````

## 1-3. 여러 개의 plot을 그리는 방법 (subplot)
`subplot(row, column, index)` 
* `index`는 1부터 시작함

````python
plt.subplot(2, 1, 1)  # 2개의 행을 만들고, 하나의 열을 만들어서 첫 번째 plot에 표시
plt.plot(data2)

plt.subplot(2, 1, 2)
plt.plot(data1)

plt.show()
````

위와 똑같은 그래프를 그리지만 다른 표현
* 콤마(,)를 생략하고 그릴 수 있음

````python
plt.subplot(211)
plt.plot(data2)

plt.subplot(212)
plt.plot(data1)

plt.show()
````

````python
plt.subplot(121)
plt.plot(data2)

plt.subplot(122)
plt.plot(data1)

plt.show()
````

## 1-4. 여러 plot을 그리는 방법(subplots) -s가 더 붙음
* `figure, axes = plt.subplots(행의 개수, 열의 개수)`
* `figure` : 캔버스 설정을 할 수 있는 객체
* `axes` : `ax` (그래프를 그릴 수 있는 공간)이 여러 개 있는 것 (`ax`의 배열)

````python
fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data1)
axes[0, 1].plot(data1 * data2)
axes[0, 2].plot(data1 ** 3)

axes[1, 0].plot(data2 % 10)
axes[1, 1].plot(-1 * data1)
axes[1, 2].plot(data1 // 20)

plt.tight_layout()  # 여러 개의 subplots 그래프를 그릴 때 캔버스를 꽉 채우도록 하는 옵션 (선택)
plt.show()
````

## 2-1. title
그래프의 제목

````python
plt.plot(data1)
plt.plot(data2)

# 그래프의 제목 설정하기
# plt.title('여기는 타이틀 입니다.')

# 폰트 사이즈 설정하기
plt.title('여기는 타이틀 입니다.', fontsize=20)
plt.show()
````

## 2-2. X, Y 축 Label 설정

````python
plt.plot(data1)
plt.plot(data2)

plt.title('label 설정 예제 입니다.')

# x축 label 설정
plt.xlabel('x축 입니다.', fontsize=20)

#y축 label 설정
plt.ylabel('y축 입니다.')

plt.show()
````

## 2-3. tick 설정하기 (rotation)

````python
plt.plot(data1 * 2)
plt.plot(data1 ** 2)
plt.plot(np.log(data1))

# 타이틀 & font 설정
plt.title('x, y tick rotation')

# x축, y축 label 부여
plt.xlabel('x축')
plt.ylabel('y축')

# x tick, y tick 설정
plt.xticks(rotation=90)  # rotation : 텍스트가 돌아가는 각도
plt.yticks(rotation=30)

plt.show()
````

## 2-4. 범례 (legend) 설정
범례를 따로 넣어주기

````python
plt.plot(data1 * 2)
plt.plot(data1 ** 2)
plt.plot(np.log(data1))

# 타이틀 & font 설정
plt.title('x, y tick rotation')

# x축, y축 label 부여
plt.xlabel('x축')
plt.ylabel('y축')

# x tick, y tick 설정
plt.xticks(rotation=90)  # rotation : 텍스트가 돌아가는 각도
plt.yticks(rotation=30)

# legend(범례) 설정
plt.legend(['data1 * 2', 'data1 ** 2', 'log(data1'])

plt.show()
````

plot과 함께 label 넣어서 범례 설정하기

````python
plt.plot(data1 * 2, label='a')  # 그래프를 그리면서 label에 값을 넣어주면 범례 이름이 됨
plt.plot(data1 ** 2, label='b')
plt.plot(np.log(data1), label='c')

# 타이틀 & font 설정
plt.title('x, y tick rotation')

# x축, y축 label 부여
plt.xlabel('x축')
plt.ylabel('y축')

# x tick, y tick 설정
plt.xticks(rotation=90)  # rotation : 텍스트가 돌아가는 각도
plt.yticks(rotation=30)

# legend(범례) 설정
plt.legend()

plt.show()
````

## 2-5. x와 y의 한계점
* tick의 한계점 설정
* 필요한 구간까지만 표시하기

````python
plt.plot(data1 * 2, label='a')  # 그래프를 그리면서 label에 값을 넣어주면 범례 이름이 됨
plt.plot(data1 ** 2, label='b')
plt.plot(np.log(data1), label='c')

# 타이틀 & font 설정
plt.title('x, y tick rotation')

# x축, y축 label 부여
plt.xlabel('x축')
plt.ylabel('y축')

# x tick, y tick 설정
plt.xticks(rotation=90)  # rotation : 텍스트가 돌아가는 각도
plt.yticks(rotation=30)

# legend(범례) 설정
plt.legend()

# x, y의 limit 설정 (특정 구간만 볼 수 있음)
plt.xlim(0, 10)
plt.ylim(-10, 100)

plt.show()
````

## 2-6. 스타일 세부 설정 marker, line, color 설정

### marker
**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker


````python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', markersize=5)
plt.plot(data, data * 2 - 10, marker='H', markersize=5)
plt.plot(data, data * 2 - 20, marker='D', markersize=5)
plt.plot(data, data * 2 - 30, marker='*', markersize=10)
````

### line

**line의 종류**
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style

````python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='')  # 선이 안 보임
plt.plot(data, data * 2 - 10, marker='H', linestyle='-')  # 실선
plt.plot(data, data * 2 - 20, marker='D', linestyle='--') # 짧은 실선
plt.plot(data, data * 2 - 30, marker='*', linestyle='-.') # 선-점
plt.plot(data, data * 2 - 40, marker='^', linestyle=':')   # 점선

plt.show()
````

### color , 투명도(alpha)

**color의 종류**
* 'b'	blue
* 'g'	green
* 'r'	red
* 'c'	cyan
* 'm'	magenta
* 'y'	yellow
* 'k'	black
* 'w'	white

````python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='', color='b')  
plt.plot(data, data * 2 - 10, marker='H', linestyle='-', color='c') 
plt.plot(data, data * 2 - 20, marker='D', linestyle='--', color='y') 
plt.plot(data, data * 2 - 30, marker='*', linestyle='-.', color='k') 
plt.plot(data, data * 2 - 40, marker='^', linestyle=':')   

plt.show()
````

alpha를 사용하면 투명도 조절이 가능
* 0 ~ 1 사이로 설정
* 0에 가까울수록 투명, 1에 가까울수록 불투명

````python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='', color='b', alpha=0.9)  
plt.plot(data, data * 2 - 10, marker='H', linestyle='-', color='c', alpha=0.7) 
plt.plot(data, data * 2 - 20, marker='D', linestyle='--', color='y', alpha=0.5) 
plt.plot(data, data * 2 - 30, marker='*', linestyle='-.', color='k', alpha=0.3) 
plt.plot(data, data * 2 - 40, marker='^', linestyle=':', alpha=0.1)   

plt.show()
````
