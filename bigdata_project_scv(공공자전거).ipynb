```python
# read_csv dep, rent total 2019~2020
dep_total_2019 = pd.read_csv('/content/drive/MyDrive/source_최종/dep_total_2019.csv')
dep_total_2020 = pd.read_csv('/content/drive/MyDrive/source_최종/dep_total_2020.csv')
rent_total_2019 = pd.read_csv('/content/drive/MyDrive/source_최종/rent_total_2019.csv')
rent_total_2020 = pd.read_csv('/content/drive/MyDrive/source_최종/rent_total_2020.csv')
```

## 그래프 라벨 저장
- gu_list: 자치구 시리즈(한국어)
- months: 월 리스트(영어)
- months_num: 월 리스트(숫자)

```python
# 라벨 자치구를 시리즈로 저장
gu_list = dep_total_2019['자치구']
# 라벨 월 리스트로 저장
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
months_num = [x for x in range(1, 13)]
# color[0] - 2019, color[1] - 2020
color = ['#51c4d3', '#e84545']
```

# 2019 전체 자치구 월별(plot)

## 2019년 전체 자치구 배치, 대여, 이용률 합계 저장

```python
# 2019년 월별 전체 자치구 배치 합계 시리즈
dep_mon_list_19 = dep_total_2019.iloc[:, 1:-1].sum()
# dep_mon_list_19
```

```python
# 2019년 월별 전체 자치구 대여 합계 시리즈
rent_mon_list_19 = rent_total_2019.iloc[:, 1:-1].sum()
# rent_mon_list_19
```

```python
# 2019년 월별 전체 자치구 이용률 합계 시리즈
# 총 이용률 = 총 대여수 / 총 배차수
avg_mon_list_19 = round(rent_mon_list_19 / dep_mon_list_19, 2)
# avg_mon_list_19
```

# 2019, 2020년도 구별 월평균 이용률(bar)

## 2019년 각 구 월별 이용률 평균
- 각 구 월별 이용률 평균 = 각 구 월별 대여수 / 각 구 월별 배차수

```python
# 2019 각 구 월별 이용량 평균
gu_use_mon_19 = pd.concat([gu_list, round(rent_total_2019.iloc[:, 1:-1] / dep_total_2019.iloc[:, 1:-1],2)], axis=1)
# gu_use_mon_19
```

```python
gu_use_19 = round(gu_use_mon_19.iloc[:].sum(axis=1) / 12, 2)
# gu_use_19
```

```python
plt.bar(gu_use_mon_19.loc[:]['자치구'], gu_use_19.sort_values(), color=color[0])
plt.xticks(rotation=45)
plt.title('2019년 각 구의 연평균 이용량')
plt.xlabel('자치구')
plt.ylabel('이용량')
plt.rc('figure', figsize=(20, 6))
plt.show()
```

## 2020년 각 구 월별 이용량 평균

```python
# 2020년 각 구 월별 이용량 평균
gu_use_mon_20 = pd.concat([gu_list, round(rent_total_2020.iloc[:, 1:-1] / dep_total_2020.iloc[:, 1:-1],2)], axis=1)
# gu_use_mon_20
```

```python
gu_use_20 = round(gu_use_mon_20.iloc[:].sum(axis=1) / 12, 2)
# gu_use_20
```

```python
plt.bar(gu_use_mon_20.loc[:]['자치구'], gu_use_20.sort_values(), color=color[1])
plt.xticks(rotation=45)
plt.title('2020년 각 구의 연평균 이용량')
plt.xlabel('자치구')
plt.ylabel('이용량')
plt.rc('figure', figsize=(20, 6))
plt.show()
```

```python
m1_19=folium.Map(
    location=[37.566345, 126.977893], 
    tiles='cartodbpositron', zoom_start=11
)

m1_20=folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron', zoom_start=11
)

m2_19=folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron', zoom_start=11
)

m2_20=folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron', zoom_start=11
)
m3=folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron', zoom_start=11)
```

```python
dif=pd.DataFrame(gu_use_20 - gu_use_19, columns=['차이'])
dif=dif.assign(자치구=list(np.array(gu_use_mon_20.loc[:]['자치구'])))
dif
```

```python
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=rent_19,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5
).add_to(m1_19)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=rent_20,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5
).add_to(m1_20)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=dep_19,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5
).add_to(m2_19)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=dep_20,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5
).add_to(m2_20)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=dif,
    columns=['자치구','차이'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5
).add_to(m3)
```
