```python
import numpy as np
from scipy.interpolate import splprep, splev

import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch

N = 400
t = np.linspace(0, 2 * np.pi, N)
r = 0.5 + np.cos(t)
x, y = r * np.cos(t), r * np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4])    # plot([x], y, [fmt-format], *, data=None, **kwargs-keyword argument) -> []는 생략 가능
plt.show()                # plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```

```python
plt.plot([1, 7, 9, 16])    # x축을 생략하고 y축만 나타낼 수 있음
plt.show()
```

```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([1, 7, 9, 16])
a1 , a2
```

```python
plt.plot(a1, a2)    # a1을 x축, a2를 y축으로 입력함 
# -> 기존 a1에 x축이 생략되어 있기 때문에, 별도의 값 생성되지 않음
```

```python
plt.plot(['a', 'b', 'c', 'd'], a2)
plt.show()
```

```python
a3 = ['a', 'b', 'c', 'd']

plt.plot(a3, a2)
plt.figure(figsize=(2,8))    ## figsize=(x,y)
plt.show()
```

```python
plt.plot(a3, a2, color='g', marker='P', linestyle='-.')   # == plt.plot(a3, a2, 'gp-.')
plt.show()     
```

marker 종류 s -> square, ^ / > -> triangle , o -> circle , P -> plus, p -> pantagone                                                   
https://matplotlib.org/stable/tutorials/colors/colors.html#sphx-glr-tutorials-colors-colors-py 에서 color 확인할 수 있음                                                                                        https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=pyplot%20plot#matplotlib.pyplot.plot -> 추가 도움말

```python
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
x, y1, y2, y3
```

```python
plt.plot(x, y1, 'rp-', markersize=8)  
plt.plot(x, y2, 'k^:', markersize=8)  
plt.plot(x, y3, 'b-.', markersize=8)  
plt.title('my picture')
plt.legend(['y1', 'y2', 'y3'])
plt.show()  
```

```python
plt.plot(x, y1, 'rp-', x, y2, 'k^:', x, y3, 'b-.', markersize=8)    
plt.title('my picture')
plt.grid(which='major', axis='both', color='gray', alpha=0.8)
plt.legend(['y1', 'y2', 'y3'])
plt.show()  
```

```python
plt.bar(a1, a2)    # 이 때 x축과 y축의 값이 같아야 함(ex.배열의 원소 수)
```

```python
color = ['lightgreen', 'lightpink', 'skyblue', 'orange']
hobby = ['여행', '독서', '영화', '음악']

numbers = [10, 5, 15, 18]

plt.figure(figsize=(7, 7))

plt.pie(numbers, labels=hobby, colors=color, autopct='%.1f%%')    # '%.1f%%' -> 소수점 이하 1자리 숫자 나타냄

plt.axis('equal')
plt.title('취미 비율')

plt.show()
```
